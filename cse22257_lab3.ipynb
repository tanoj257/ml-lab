```python
import pandas as pd
import numpy as np

# To Load the excel data
def load_data(file_path, sheet_name):
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    return df


# Function to calculate dimensionality
def dimensions(A):
     # shape is an attribute that provides the dimensions of an array present in numpy library
     # it returns a tuple representing the dimensions
    dimensionality = A.shape
    print(f"Dimensionality of the vector space: {dimensionality}")

   
# Function to count the number of vectors
def count_vectors(A):
     # A.shape[0] gives the number of rows
        #A.shape[1] gives the number of columns
    num_vectors = A.shape[0]
    print(f"Number of vectors in this vector space: {num_vectors}")

       
# Function to find the rank of Matrix A
def rank_of_matrix(A):
    # numpy.linalg.matrix_rank() gives the number of linearly independent rows or columns
    rank_A = np.linalg.matrix_rank(A)
    print(f"Rank of Matrix A: {rank_A}")

# Function to compute pseudo-inverse and find costs
def compute_costs(A, C):
    # np.linalg.pinv() is used to get the inverse of A
    A_pseudo_inv = np.linalg.pinv(A)
   
    # X is used to get the cost of the pseudo inverse
    #dot() performs dot product
    # this function call performs the matrix multiplication
    X = np.dot(A_pseudo_inv, C)
    return X


# Function to mark customers as RICH or POOR and build a classifier
def classify_customers(df):
   
    # lambda function to categorize
    df['Class'] = df['Payment (Rs)'].apply(lambda x: 'RICH' if x > 200 else 'POOR')
    print(df)
   
# main function
def main():
    # Load the data
    file_path = r"C:\Users\year3\Downloads\Lab Session Data (1).xlsx"
    sheet_name = 'Purchase data'
    df = load_data(file_path, sheet_name)

    # Selection of specific columns
    columns_to_print = df[['Customer', 'Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)', 'Payment (Rs)']]
    print("Selected columns:")
    print(columns_to_print)
   
    # Create matrices A and C
    A = df[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].values
    C = df[['Payment (Rs)']].values
   
    # Print matrices
    print("\nMatrix A:")
    print(A)
    print("\nMatrix C:")
    print(C)
   
    # Perform calculations
    dimensions(A)
    count_vectors(A)
    rank_of_matrix(A)
    X = compute_costs(A, C)
    print("\nCost of each product available for sale:")
    print(X)
   
    # Classify customers
    classify_customers(df)

# Execute the main function
if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt

def load_data(file_path, sheet_name):
    # Dataframe to store the data
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    return df


 #  Function to calculate the mean and variance of the Price data in column D.
def calculate_mean_variance(df):
    # accessing the coloumn based on the name
    price_data = df['Price']
    mean_price = statistics.mean(price_data)
    variance_price = statistics.variance(price_data)
    print(f"Mean of Price data: {mean_price}")
    print(f"Variance of Price data: {variance_price}")
    return mean_price


#  To get the mean of days which are wednesdays
def sample_mean_wednesdays(df, population_mean):  
    # to get only rows with wednesday
    wednesdays = df[df['Day'] == 'Wed']
    # to access price col in rows with wednesday
    wednesday_prices = wednesdays['Price']
    # mean of wednesday prices
    mean_wednesday_price = statistics.mean(wednesday_prices)
    print(f"Sample mean price for Wednesdays: {mean_wednesday_price}")
    print(f"Comparison with population mean: {mean_wednesday_price - population_mean}")

   
    # To get the price mean of April month
def sample_mean_april(df, population_mean):
     # to get only rows with April month
    april = df[df['Month'] == 'Apr']
      # to access price col in rows with April month
    april_prices = april['Price']
    # mean of April prices
    mean_april_price = statistics.mean(april_prices)
    print(f"Sample mean price for April: {mean_april_price}")
    print(f"Comparison with population mean: {mean_april_price - population_mean}")

   
 # Function to find the probability of making loss over the stock on Chg%
def probability_of_loss(df):
    # to get the Chg col
    chg_percent = df['Chg%']
    """chg_percent.apply(lambda x: x < 0) is the lambda function which checks each
       value in the Chg% col is less than 0 resulting in a boolean Series where True
       corresponds to a loss. """
  #  counts the number of True values
    loss_count = sum(chg_percent.apply(lambda x: x < 0))
   
  # counts the probability by dividing count of losses by total
    loss_probability = loss_count / len(chg_percent)
    print(f"Probability of making a loss: {loss_probability}")
    return loss_probability


# To get the probability of making profit on wed
def probability_of_profit_wednesday(df):
    # filter for wed
    wednesdays = df[df['Day'] == 'Wed']
    # to get the percentage cols for retrived rows
    wednesday_chg_percent = wednesdays['Chg%']
    # to count the profit
    profit_count = sum( wednesday_chg_percent.apply(lambda x: x > 0))
    # to calculate the probability
    profit_probability_wednesday = profit_count / len(wednesday_chg_percent)
    print(f"Probability of making a profit on Wednesday: {profit_probability_wednesday}")
    return profit_probability_wednesday

# Calculate the conditional probability of making a profit, given that today is Wednesday.
def conditional_probability_of_profit(df, profit_probability_wednesday):
    chg_percent = df['Chg%']
    profit_count = sum(chg_percent.apply(lambda x: x > 0))
    profit_probability = profit_count / len(chg_percent)
    # conditional probabliity
    conditional_probability = profit_probability_wednesday / profit_probability
    if profit_probability > 0:
        print(f"Conditional probability of making a profit, given that today is Wednesday: {conditional_probability}")
    else:
         0
    return conditional_probability

#scatter plot of Chg% data against the day of the week.
def scatter_plot_chg_day(df):
    # converts the Date column of the DataFrame df to a datetime format.
    df['Date'] = pd.to_datetime(df['Date'])
   
    df['Day'] = df['Date'].dt.day_name()
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Day'], df['Chg%'], alpha=0.5)
    plt.xlabel('Day of the Week')
    plt.ylabel('Chg%')
    plt.title('Scatter Plot of Chg% Data Against Day of the Week')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

   
def main():
    file_path =  r"C:\Users\year3\Downloads\Lab Session Data (1).xlsx"
    sheet_name = 'IRCTC Stock Price'
   
    # To Load the data
    df = load_data(file_path, sheet_name)
   
    # Calculate mean and variance
    population_mean = calculate_mean_variance(df)
   
    # Calculate sample mean for Wednesdays
    sample_mean_wednesdays(df, population_mean)
   
    # Calculate sample mean for April
    sample_mean_april(df, population_mean)
   
    probability_of_loss(df)
   
    probability_of_profit_wednesday(df)
    profit_probability_wednesday = probability_of_profit_wednesday(df)
   
    conditional_probability_of_profit(df, profit_probability_wednesday)
   
    scatter_plot_chg_day(df)
# Execute the main function
if __name__ == "__main__":
    main()
```
